```python
from LCS_DP import DynamicLCS
from DNC import DNC_LCS
from time import perf_counter
from matplotlib import pyplot as plt
```

```python
a = 'GTCGTTCGGAATGCCGTTCTGTAAA'
b = 'ACCGGTCGAGTGCGCGGACGGCCGAA'
```

```python

input_gen = ((a[:i:], b[:i:]) for i in range(len(a)-1) if i > 0)
x = []
y_Dynamic = []
y_DNC = []

for i in input_gen:
    x.append(len(i[0])+len(i[1]))
    start = perf_counter()
    DynamicLCS(i[0], i[1])
    end = perf_counter()
    y_Dynamic.append(end - start)
    start = perf_counter()
    DNC_LCS(i[0], i[1])
    end = perf_counter()
    y_DNC.append(end - start)

plt.rcParams["figure.figsize"] = (10, 10)
plt.plot(x, y_Dynamic)
plt.plot(x, y_DNC)
plt.show()
```
